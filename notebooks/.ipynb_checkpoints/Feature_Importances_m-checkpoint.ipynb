```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Titanic data + random columns

```python
'''
output
- titanic, shape(1309, 14), 4 int64 columns, 7 object columns, 3 float64 columns
------------------------------- Data Info Table -------------------------------
         pclass  survived                           name     sex   age  sibsp  \
example       1         1  Allen, Miss. Elisabeth Walton  female  29.0      0   

         parch ticket      fare cabin embarked boat  body     home.dest  
example      0  24160  211.3375    B5        S    2   NaN  St Louis, MO  
------------------------------- Data Info Table -------------------------------

'''
titanic = pd.read_csv("titanic3.csv")
```

```python
'''
input
- titanic, shape(1309, 14), 4 int64 columns, 7 object columns, 3 float64 columns
output
- titanic, shape(1309, 16), 4 int64 columns, 7 object columns, 4 float64 columns, 1 int32 columns
------------------------------- Data Info Table -------------------------------
         pclass  survived                           name     sex   age  sibsp  \
example       1         1  Allen, Miss. Elisabeth Walton  female  29.0      0   

         parch ticket      fare cabin embarked boat  body     home.dest  \
example      0  24160  211.3375    B5        S    2   NaN  St Louis, MO   

         random_cat  random_num  
example           2     -1.1106  
------------------------------- Data Info Table -------------------------------

'''
rng = np.random.RandomState(42)
titanic['random_cat'] = rng.randint(3, size=titanic.shape[0])
titanic['random_num'] = rng.randn(titanic.shape[0])
```

```python
'''
input
- titanic, shape(1309, 16), 4 int64 columns, 7 object columns, 4 float64 columns, 1 int32 columns
'''
titanic.describe()
```

```python
'''
input
- titanic, shape(1309, 16), 4 int64 columns, 7 object columns, 4 float64 columns, 1 int32 columns
'''
titanic.head(10)
```

```python
'''
input
- titanic, shape(1309, 16), 4 int64 columns, 7 object columns, 4 float64 columns, 1 int32 columns
output
- data, shape(1309, 9), 3 int64 columns, 2 object columns, 1 int32 columns, 3 float64 columns
------------------------------- Data Info Table -------------------------------
         pclass     sex embarked  random_cat   age  sibsp  parch      fare  \
example       1  female        S           2  29.0      0      0  211.3375   

         random_num  
example     -1.1106  
------------------------------- Data Info Table -------------------------------

- labels, shape(1309,) of int64
          mean: 0.3820, variance: 0.2361, range: [0, 1]
- categorical_columns, <class 'list'>, ['pclass', 'sex', 'embarked', 'random_cat']
- numerical_columns, <class 'list'>, ['age', 'sibsp', 'parch', 'fare', 'random_num']
'''
categorical_columns = ['pclass', 'sex', 'embarked', 'random_cat']
numerical_columns = ['age', 'sibsp', 'parch', 'fare', 'random_num']

data = titanic[categorical_columns + numerical_columns]
labels = titanic['survived']
```

```python
'''
input
- data, shape(1309, 9), 3 int64 columns, 2 object columns, 1 int32 columns, 3 float64 columns
- labels, shape(1309,) of int64
output
- X_train, shape(981, 9), 3 int64 columns, 2 object columns, 1 int32 columns, 3 float64 columns
------------------------------- Data Info Table -------------------------------
         pclass     sex embarked  random_cat  age  sibsp  parch    fare  \
example       3  female        Q           1  NaN      0      0  7.7333   

         random_num  
example     -0.4364  
------------------------------- Data Info Table -------------------------------

- y_train, shape(981,) of int64
           mean: 0.3823, variance: 0.2361, range: [0, 1]
- X_test, shape(328, 9), 3 int64 columns, 2 object columns, 1 int32 columns, 3 float64 columns
------------------------------- Data Info Table -------------------------------
         pclass   sex embarked  random_cat  age  sibsp  parch   fare  \
example       3  male        S           2  NaN      8      2  69.55   

         random_num  
example      0.9192  
------------------------------- Data Info Table -------------------------------

- y_test, shape(328,) of int64
          mean: 0.3811, variance: 0.2359, range: [0, 1]
'''
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(
    data, labels, stratify=labels, random_state=42)
```

## Building a Feature Engineering Pipeline

```python
'''
input
- categorical_columns, <class 'list'>
- numerical_columns, <class 'list'>
output
- rf, <class 'sklearn.pipeline.Pipeline'>, Pipeline(steps=[('preprocess',
                 ColumnTransformer(transformers=[('cat',
                                                  Pipeline(steps=[('imputer',
                                                                   SimpleImputer(fill_value='missing',
                                                                                 strategy='constant')),
                                                                  ('onehot',
                                                                   OneHotEncoder(handle_unknown='ignore'))]),
                                                  ['pclass', 'sex', 'embarked',
                                                   'random_cat']),
                                                 ('num',
                                                  Pipeline(steps=[('imputer',
                                                                   SimpleImputer())]),
                                                  ['age', 'sibsp', 'parch',
                                                   'fare', 'random_num'])])),
                ('classifier', RandomForestClassifier(random_state=42))])
- permutation_importance, <class 'function'>, <function permutation_importance at 0x00000178CC5BE6A8>
'''
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
from sklearn.inspection import permutation_importance
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder

categorical_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore')),
])
numerical_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    # no scaler needed for random forests
])

preprocessing = ColumnTransformer(
    [('cat', categorical_pipe, categorical_columns),
     ('num', numerical_pipe, numerical_columns)])

rf = Pipeline([
    ('preprocess', preprocessing),
    ('classifier', RandomForestClassifier(n_estimators=100,
                                          min_samples_leaf=1,
                                          random_state=42))
])
```

```python
'''
input
- rf, <class 'sklearn.pipeline.Pipeline'>
- X_train, shape(981, 9), 3 int64 columns, 2 object columns, 1 int32 columns, 3 float64 columns
- y_train, shape(981,) of int64
output
- X_train, shape(981, 9), 3 int64 columns, 2 object columns, 1 int32 columns, 3 float64 columns, no change in the cell
------------------------------- Data Info Table -------------------------------
         pclass     sex embarked  random_cat  age  sibsp  parch    fare  \
example       3  female        Q           1  NaN      0      0  7.7333   

         random_num  
example     -0.4364  
------------------------------- Data Info Table -------------------------------

- y_train, shape(981,) of int64
           mean: 0.3823, variance: 0.2361, range: [0, 1]
- rf, <class 'sklearn.pipeline.Pipeline'>, Pipeline(steps=[('preprocess',
                 ColumnTransformer(transformers=[('cat',
                                                  Pipeline(steps=[('imputer',
                                                                   SimpleImputer(fill_value='missing',
                                                                                 strategy='constant')),
                                                                  ('onehot',
                                                                   OneHotEncoder(handle_unknown='ignore'))]),
                                                  ['pclass', 'sex', 'embarked',
                                                   'random_cat']),
                                                 ('num',
                                                  Pipeline(steps=[('imputer',
                                                                   SimpleImputer())]),
                                                  ['age', 'sibsp', 'parch',
                                                   'fare', 'random_num'])])),
                ('classifier', RandomForestClassifier(random_state=42))])
'''
rf.fit(X_train, y_train)
```

```python
'''
input
- X_train, shape(981, 9), 3 int64 columns, 2 object columns, 1 int32 columns, 3 float64 columns
- y_train, shape(981,) of int64
- rf, <class 'sklearn.pipeline.Pipeline'>
- X_test, shape(328, 9), 3 int64 columns, 2 object columns, 1 int32 columns, 3 float64 columns
- y_test, shape(328,) of int64
'''
print("RF train accuracy: %0.3f" % rf.score(X_train, y_train))
print("RF test accuracy: %0.3f" % rf.score(X_test, y_test))
```

## Tree-based Feature Importances

```python
'''
input
- rf, <class 'sklearn.pipeline.Pipeline'>
- categorical_columns, <class 'list'>
- numerical_columns, <class 'list'>
'''
ohe = (rf.named_steps['preprocess']
         .named_transformers_['cat']
         .named_steps['onehot'])
feature_names = []
for col, cats in zip(categorical_columns, ohe.categories_):
    for cat in cats:
        feature_names.append("{}_{}".format(col, cat))
feature_names = np.array(feature_names + numerical_columns)

tree_feature_importances = (
    rf.named_steps['classifier'].feature_importances_)
sorted_idx = tree_feature_importances.argsort()

y_ticks = np.arange(0, len(feature_names))
_, ax = plt.subplots(figsize=(12, 8))
ax.barh(y_ticks, tree_feature_importances[sorted_idx])
ax.set_yticklabels(feature_names[sorted_idx])
ax.set_yticks(y_ticks)
ax.set_title("Random Forest Feature Importances");
```

## Permutation-based Feature Importances (training set)

```python
'''
input
- rf, <class 'sklearn.pipeline.Pipeline'>
- X_train, shape(981, 9), 3 int64 columns, 2 object columns, 1 int32 columns, 3 float64 columns
- y_train, shape(981,) of int64
- permutation_importance, <class 'function'>
- X_test, shape(328, 9), 3 int64 columns, 2 object columns, 1 int32 columns, 3 float64 columns
output
- rf, <class 'sklearn.pipeline.Pipeline'>, Pipeline(steps=[('preprocess',
                 ColumnTransformer(transformers=[('cat',
                                                  Pipeline(steps=[('imputer',
                                                                   SimpleImputer(fill_value='missing',
                                                                                 strategy='constant')),
                                                                  ('onehot',
                                                                   OneHotEncoder(handle_unknown='ignore'))]),
                                                  ['pclass', 'sex', 'embarked',
                                                   'random_cat']),
                                                 ('num',
                                                  Pipeline(steps=[('imputer',
                                                                   SimpleImputer())]),
                                                  ['age', 'sibsp', 'parch',
                                                   'fare', 'random_num'])])),
                ('classifier', RandomForestClassifier(random_state=42))])
- X_train, shape(981, 9), 3 int64 columns, 2 object columns, 1 int32 columns, 3 float64 columns, no change in the cell
------------------------------- Data Info Table -------------------------------
         pclass     sex embarked  random_cat  age  sibsp  parch    fare  \
example       3  female        Q           1  NaN      0      0  7.7333   

         random_num  
example     -0.4364  
------------------------------- Data Info Table -------------------------------

- y_train, shape(981,) of int64
           mean: 0.3823, variance: 0.2361, range: [0, 1]
'''
permute_importance = permutation_importance(rf, X_train, y_train, n_repeats=30,
                                            random_state=42).importances

permute_importance_mean = np.mean(permute_importance, axis=-1)
sorted_idx = permute_importance_mean.argsort()

_, ax = plt.subplots(figsize=(12, 8))
ax.boxplot(permute_importance[sorted_idx].T,
           vert=False, labels=X_test.columns[sorted_idx])
ax.vlines(0, 0, X_test.shape[1] + 1, linestyles='dashed', alpha=0.5)
ax.set_xlabel("baseline score - score on permutated variable")
ax.set_title("Permutation Importances (training set)");
```

## Permutation-based Feature Importances (test set)

```python
'''
input
- rf, <class 'sklearn.pipeline.Pipeline'>
- X_test, shape(328, 9), 3 int64 columns, 2 object columns, 1 int32 columns, 3 float64 columns
- y_test, shape(328,) of int64
- permutation_importance, <class 'function'>
output
- rf, <class 'sklearn.pipeline.Pipeline'>, Pipeline(steps=[('preprocess',
                 ColumnTransformer(transformers=[('cat',
                                                  Pipeline(steps=[('imputer',
                                                                   SimpleImputer(fill_value='missing',
                                                                                 strategy='constant')),
                                                                  ('onehot',
                                                                   OneHotEncoder(handle_unknown='ignore'))]),
                                                  ['pclass', 'sex', 'embarked',
                                                   'random_cat']),
                                                 ('num',
                                                  Pipeline(steps=[('imputer',
                                                                   SimpleImputer())]),
                                                  ['age', 'sibsp', 'parch',
                                                   'fare', 'random_num'])])),
                ('classifier', RandomForestClassifier(random_state=42))])
- X_test, shape(328, 9), 3 int64 columns, 2 object columns, 1 int32 columns, 3 float64 columns, no change in the cell
------------------------------- Data Info Table -------------------------------
         pclass   sex embarked  random_cat  age  sibsp  parch   fare  \
example       3  male        S           2  NaN      8      2  69.55   

         random_num  
example      0.9192  
------------------------------- Data Info Table -------------------------------

- y_test, shape(328,) of int64
          mean: 0.3811, variance: 0.2359, range: [0, 1]
- permute_importance, shape(9, 30) of float64
                      mean: 0.0365, variance: 0.0047, range: [-0.03658536585365857, 0.25914634146341453]
'''
permute_importance = permutation_importance(rf, X_test, y_test, n_repeats=30,
                                            random_state=42).importances
permute_importance_mean = np.mean(permute_importance, axis=-1)
sorted_idx = permute_importance_mean.argsort()

_, ax = plt.subplots(figsize=(12, 8))
ax.boxplot(permute_importance[sorted_idx].T,
           vert=False, labels=X_test.columns[sorted_idx])
ax.vlines(0, 0, X_test.shape[1] + 1, linestyles='dashed', alpha=0.5)
ax.set_xlabel("baseline score - score on permutated variable")
ax.set_title("Permutation Importances (test set)");
```
